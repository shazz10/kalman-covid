```python
import h2o
from h2o.estimators import H2ORandomForestEstimator
from h2o.estimators.glm import H2OGeneralizedLinearEstimator
from h2o.grid.grid_search import H2OGridSearch
h2o.init(min_mem_size='2G')
import numpy as np
from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
train = t[t["date"]<='2020-05-02']
```

```python
test = t[t["date"]>="2020-05-03"]
```

```python
train = train.drop(["kalman_prediction"],axis=1)
```

```python
test = test.drop(["kalman_prediction"],axis=1)
```

```python
train = train[train["region"]=="India_Maharashtra"]
test = test[test["region"]=="India_Maharashtra"]
```

```python
train_h20 = h2o.H2OFrame(train)

training_columns = ["Growth in 1 day",
                    "Growth in 3 days","Growth in 5 days",
                      "Growth in 7 days","Growth rate for 1 day",
                     "Growth rate for 3 days","Growth rate for 5 days",
                      "Growth rate for 7 days","Cases 1 day ago",
             
       'mintempC', 'maxtempC','humidity'
          ]               
# Output parameter train against input parameters
response_column = 'Confirmed Cases'

model = H2ORandomForestEstimator(ntrees=300, max_depth=12)
model.train(x=training_columns, y=response_column, training_frame=train_h20)
#model = H2ORandomForestEstimator(ntrees=300, max_depth=12)
# model = H2OGeneralizedLinearEstimator()
# model.train(x=training_columns, y=response_column, training_frame=train_h20)
```

```python
feature_importance = model.varimp(True).iloc[:,:]
feature_importance=feature_importance.set_index("variable")
feature_importance
```

```python
del feature_importance["relative_importance"]
feature_importance["scaled_importance"] = round(feature_importance["scaled_importance"],2)
feature_importance["percentage"] = round(feature_importance["percentage"],2)
```

```python
ax=feature_importance.plot(kind="bar",figsize=(20,10))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
fig = ax.get_figure()
fig.savefig("feature_importance"+".png")
```

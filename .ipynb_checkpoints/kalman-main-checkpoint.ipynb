# CoronaVirus Prediction

```python
# Get data from Github
import numpy as np
from math import sqrt
from sklearn.metrics import mean_squared_error
import pandas as pd
```

```python
df = pd.read_csv("covid_19_india9may.csv")
df.head(5)
```

```python
df["State/UnionTerritory"].unique()
```

```python
data_c={}
data_d={}
data_r={}
for state in df["State/UnionTerritory"].unique():
    new1={}
    new2={}
    new3={}
    df_cur = df.loc[df['State/UnionTerritory'] == state]
    for i in range(df_cur.shape[0]):
        new1[df_cur.iloc[i]["Date"]] = df_cur.iloc[i]["Confirmed"]
        new2[df_cur.iloc[i]["Date"]] = df_cur.iloc[i]["Deaths"]
        new3[df_cur.iloc[i]["Date"]] = df_cur.iloc[i]["Cured"]
    data_c[state]=new1
    data_d[state]=new2
    data_r[state]=new3
```

```python
dates = data_c["Kerala"]
for state in df["State/UnionTerritory"].unique():
    for d in dates.keys():
        if d not in data_c[state].keys():
            data_c[state][d]=0
            data_r[state][d]=0
            data_d[state][d]=0
```

```python
df2 = pd.read_csv("population.csv")
df2.head(5)
```

```python
data_confirmed={
        "Province/State":[],
        "Country/Region":[],
        "Lat":[],
        "Long":[],
        "Population":[]
     }
for key in data_c.keys():
    if key == "Unassigned":
      continue 
    data_confirmed["Province/State"].append(key)
    data_confirmed["Country/Region"].append("India")
    data_confirmed["Lat"].append(0)
    data_confirmed["Long"].append(0)
    data_confirmed["Population"].append(df2.loc[df2["State / Union Territory"]==key]["Population"].item())

    for i in data_c[key].keys():
        if i not in data_confirmed.keys():
            data_confirmed[i]=[]
        data_confirmed[i].append(data_c[key][i])
```

```python
data_death={
        "Province/State":[],
        "Country/Region":[],
        "Lat":[],
        "Long":[],
        "Population":[]
     }
for key in data_d.keys():
    if key == "Unassigned":
      continue
    data_death["Province/State"].append(key)
    data_death["Country/Region"].append("India")
    data_death["Lat"].append(0)
    data_death["Long"].append(0)
    data_death["Population"].append(df2.loc[df2["State / Union Territory"]==key]["Population"].item())
    for i in data_d[key].keys():
        if i not in data_death.keys():
            data_death[i]=[]
        data_death[i].append(data_d[key][i])
```

```python
data_cured={
        "Province/State":[],
        "Country/Region":[],
        "Lat":[],
        "Long":[],
        "Population":[]
     }
for key in data_r.keys():
    if key == "Unassigned":
      continue
    data_cured["Province/State"].append(key)
    data_cured["Country/Region"].append("India")
    data_cured["Lat"].append(0)
    data_cured["Long"].append(0)
    data_cured["Population"].append(df2.loc[df2["State / Union Territory"]==key]["Population"].item())
    for i in data_r[key].keys():
        if i not in data_cured.keys():
            data_cured[i]=[]
        data_cured[i].append(data_r[key][i])
```

```python
len(data_confirmed["Long"])
```

```python
confirmed = pd.DataFrame(data_confirmed)
death = pd.DataFrame(data_death)
recover = pd.DataFrame(data_cured)
```

```python
confirmed['region']=confirmed['Country/Region'].map(str)+'_'+confirmed['Province/State'].map(str)
death['region']=death['Country/Region'].map(str)+'_'+death['Province/State'].map(str)
recover['region']=recover['Country/Region'].map(str)+'_'+recover['Province/State'].map(str)
confirmed.iloc[:5,:]
```

```python
df = pd.read_csv("global.csv")
df = df[df["Country/Region"]=="India"]
df["Population"] = 1352600000
df['region']=df['Country/Region'].map(str)+'_'+df['Province/State'].map(str)
df
```

```python
def create_ts(df):
    ts=df
    ts=ts.drop(['Province/State', 'Country/Region','Lat', 'Long','Population'], axis=1)
    ts.set_index('region')
    ts=ts.T
    ts.columns=ts.loc['region']
    ts=ts.drop('region')
    ts=ts.fillna(0)
    ts=ts.reindex(sorted(ts.columns), axis=1)
    return (ts)
```

```python
ts=create_ts(confirmed)
ts_d=create_ts(death)
ts_rec=create_ts(recover)
#ts = create_ts(df)
```

```python
%%R
require(pracma)
require(Metrics)
require(readr)
all<- read_csv("ts_r_global21apr.csv")
all$X1<-NULL
date<-all[,1]
date[nrow(date) + 1,1] <-all[nrow(all),1]+1
pred_all<-NULL
for (n in 2:ncol(all)-1) {
  Y<-ts(data = all[n+1], start = 1, end =nrow(all)+1)  
  sig_w<-0.01
  w<-sig_w*randn(1,100) # acceleration which denotes the fluctuation (Q/R) rnorm(100, mean = 0, sd = 1)
  sig_v<-0.01
  v<-sig_v*randn(1,100)   
  t<-0.45
  phi<-matrix(c(1,0,t,1),2,2)
  gama<-matrix(c(0.5*t^2,t),2,1)
  H<-matrix(c(1,0),1,2)
  #Kalman
  x0_0<-p0_0<-matrix(c(0,0),2,1)
  p0_0<-matrix(c(1,0,0,1),2,2)
  Q<-0.01
  R<-0.01
  X<-NULL
  X2<-NULL
  pred<-NULL
  for (i in 0:nrow(all)) {
    namp <-paste("p", i+1,"_",i, sep = "")
    assign(namp, phi%*%(get(paste("p", i,"_",i, sep = "")))%*%t(phi)+gama%*%Q%*%t(gama))
    namk <- paste("k", i+1, sep = "")
    assign(namk,get(paste("p", i+1,"_",i, sep = ""))%*%t(H)%*%(1/(H%*%get(paste("p", i+1,"_",i, sep = ""))%*%t(H)+R)))
    namx <- paste("x", i+1,"_",i, sep = "")
    assign(namx,phi%*%get(paste("x", i,"_",i, sep = "")))
    namE <- paste("E", i+1, sep = "")
    assign(namE,Y[i+1]-H%*%get(paste("x", i+1,"_",i, sep = "")))
    namx2 <- paste("x", i+1,"_",i+1, sep = "")
    assign(namx2,get(paste("x", i+1,"_",i, sep = ""))+get(paste("k", i+1, sep = ""))%*%get(paste("E", i+1, sep = "")))
    namp2 <- paste("p", i+1,"_",i+1, sep = "")
    assign(namp2,(p0_0-get(paste("k", i+1, sep = ""))%*%H)%*%get(paste("p", i+1,"_",i, sep = "")))
    X<-rbind(X,get(paste("x", i+1,"_",i,sep = ""))[1])
    X2<-rbind(X2,get(paste("x", i+1,"_",i,sep = ""))[2])
    if(i>2){
      remove(list=(paste("p", i-1,"_",i-2, sep = "")))
      remove(list=(paste("k", i-1, sep = "")))
      remove(list=(paste("E", i-1, sep = "")))
      remove(list=(paste("p", i-2,"_",i-2, sep = "")))
      remove(list=(paste("x", i-1,"_",i-2, sep = "")))
      remove(list=(paste("x", i-2,"_",i-2, sep = "")))}
  }
  pred<-NULL
  pred<-cbind(Y,X,round(X2,4))
  pred<-as.data.frame(pred)
  pred$region<-colnames(all[,n+1])
  pred$date<-date$date
  pred$actual<-rbind(0,(cbind(pred[2:nrow(pred),1])/pred[1:nrow(pred)-1,1]-1)*100)
  pred$predict<-rbind(0,(cbind(pred[2:nrow(pred),2])/pred[1:nrow(pred)-1,2]-1)*100)
  pred$pred_rate<-(pred$X/pred$Y-1)*100
  pred$X2_change<-rbind(0,(cbind(pred[2:nrow(pred),3]-pred[1:nrow(pred)-1,3])))
  pred_all<-rbind(pred_all,pred)
}
pred_all<-cbind(pred_all[,4:5],pred_all[,1:3])
names(pred_all)[5]<-"X2"
pred_all=pred_all[with( pred_all, order(region, date)), ]
pred_all<-pred_all[,3:5]
```
